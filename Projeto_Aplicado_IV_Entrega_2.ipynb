# **Introdução**

### **1.1 Contextualização**

O setor de saúde suplementar desempenha um papel fundamental na assistência à população brasileira, complementando o Sistema Único de Saúde (SUS). Segundo dados da Agência Nacional de Saúde Suplementar (ANS, 2024), mais de 50 milhões de brasileiros possuem algum tipo de plano de saúde, representando aproximadamente 24% da população. Esse número apresentou crescimento expressivo nos últimos anos, impulsionado especialmente pela pandemia de Covid-19, que aumentou a demanda por acesso rápido e especializado aos serviços de saúde.

Apesar desse crescimento, observa-se uma distribuição desigual dos planos de saúde no país, com variações significativas de acordo com fatores como faixa etária, localização geográfica e tipo de contratação. Regiões do Sudeste concentram aproximadamente 65% dos beneficiários, enquanto Norte e Nordeste apresentam taxas de adesão consideravelmente menores. Ademais, fatores econômicos e variações nos custos impactam diretamente a expansão e a sustentabilidade do setor.

### **1.2 Motivação**

A motivação para este estudo decorre da percepção de desigualdades no acesso à saúde suplementar. A distribuição desigual dos planos reflete desafios estruturais que podem comprometer a inclusão de diferentes grupos populacionais. A pandemia evidenciou essas discrepâncias e destacou a importância da acessibilidade aos serviços de saúde. Assim, identificar padrões e disparidades pode revelar lacunas que demandam ações estratégicas.

### **1.3 Justificativa**

Este estudo justifica-se pelo seu impacto social e econômico. Do ponto de vista social, entender a distribuição dos beneficiários e os fatores que influenciam o acesso aos planos de saúde pode subsidiar a formulação de políticas públicas mais eficazes, promovendo inclusão e equidade. No aspecto econômico, o setor de saúde suplementar movimenta bilhões de reais anualmente, sendo essencial compreender os padrões de crescimento e sustentabilidade financeira. Além disso, a aplicação de técnicas de Ciência de Dados e análise de séries temporais permite gerar insights relevantes para tomadores de decisão, reguladores e operadoras de planos de saúde.

### **1.4 Objetivos**

**Objetivo Geral**

Analisar a evolução do número de beneficiários de planos de saúde no Brasil entre 2019 e 2024, identificando padrões, tendências e fatores que influenciam o crescimento ou estagnação do setor, com foco em faixas etárias, tipos de contratação e distribuição geográfica por Unidade da Federação.

**Objetivos Específicos**

- Investigar o perfil demográfico dos beneficiários, considerando faixa etária, localização e tipo de contratação.

- Identificar possíveis correlações entre o crescimento de beneficiários e fatores socioeconômicos ou eventos relevantes, como a pandemia de Covid-19.

- Gerar previsões sobre a evolução do setor utilizando modelos como ARIMA, LSTM e Prophet.

- Produzir visualizações e relatórios que auxiliem operadoras e órgãos reguladores na tomada de decisão.

# **Referencial Teórico**

Este projeto tem como foco a análise do crescimento e perfil dos beneficiários de planos de saúde no Brasil, abordando conceitos como série temporal, distribuição geográfica, faixa etária e tipos de contratação (individual/familiar, coletivo empresarial ou por adesão). O objetivo principal é avaliar a evolução dos planos entre 2019 e 2024, identificando padrões e tendências relevantes para a sustentabilidade do setor. Trabalhos correlacionados (ANS, 2025) indicam que fatores como crise econômica, políticas de saúde e eventos sanitários (por exemplo, a pandemia de Covid-19) influenciam diretamente a procura por planos, enquanto pesquisas acadêmicas (SILVA; COSTA, 2021) ressaltam a importância de modelos estatísticos e de aprendizado de máquina na projeção de demanda e custos.

Estudos anteriores aplicaram diferentes técnicas para analisar esse mercado. Modelos baseados em regressão linear foram utilizados para relacionar o crescimento de beneficiários a variáveis macroeconômicas, como PIB e taxa de desemprego (MARTINS et al., 2020). Já métodos de séries temporais, como ARIMA e SARIMA, mostraram-se eficazes na previsão de flutuações no número de usuários, capturando padrões cíclicos e sazonais (SANTOS; PEREIRA, 2019). No entanto, tais modelos podem apresentar limitações quando há eventos exógenos inesperados, como crises sanitárias ou mudanças regulatórias (ANS, 2023), pois dependem de suposições sobre a estabilidade do comportamento da série ao longo do tempo.

Alternativamente, técnicas de aprendizado de máquina, como redes neurais recorrentes (RNN) e gradient boosting (XGBoost), têm sido empregadas para modelar a demanda por planos de saúde, pois conseguem integrar múltiplas variáveis de forma flexível (RIBEIRO; ALMEIDA, 2022). Apesar disso, esses métodos exigem um grande volume de dados e apresentam desafios relacionados à interpretabilidade e ao custo computacional. Trabalhos como o de LIMA et al. (2023) destacam que, embora modelos mais complexos possam oferecer maior acurácia, a transparência nos critérios de decisão é essencial para operadoras e reguladores do setor.

Diante desse contexto, o presente estudo busca aliar a robustez dos modelos de séries temporais à adaptabilidade de abordagens mais sofisticadas, considerando as especificidades da saúde suplementar no Brasil. A integração de bases de dados da ANS (2025) e de relatórios oficiais permitirá uma análise mais completa, facilitando a identificação de tendências e o suporte à tomada de decisão para operadoras, reguladores e gestores públicos. Dessa forma, espera-se contribuir para a eficiência e a equidade no acesso aos serviços de saúde suplementar no país.

# **Pipeline da Solução**

A solução proposta visa integrar e analisar os dados disponibilizados (número de beneficiários por faixa etária, UF e tipo de contratação) por meio de técnicas de série temporal e aprendizado de máquina. O fluxo de trabalho segue as etapas abaixo:

#### **1. Coleta e Integração dos Dados**
- Reunir as planilhas fornecidas (beneficiários por UF, faixa etária, tipo de contratação) em um único repositório.
- Padronizar os formatos de data, remover duplicações e corrigir inconsistências.
- Estruturar os dados em um único dataframe contendo as colunas: **"Ano/Mês", "UF", "Faixa Etária", "Tipo de Contratação", "Nº de Beneficiários"**.
- Implementar um processo de **ETL** caso necessário para transformação e armazenamento intermediário dos dados.

#### **2. Limpeza e Preparação**
- Tratar valores ausentes ou inválidos, utilizando diferentes abordagens dependendo do caso (exclusão se superior a 10% dos dados, interpolação se inferior a esse limite, substituição pela média ou mediana quando apropriado).
- Criar colunas derivadas para facilitar a análise temporal (exemplo: **mês consolidado, taxa de variação mensal**).
- Normalizar ou padronizar variáveis caso necessário para compatibilidade com os modelos de aprendizado de máquina.

#### **3. Análise Exploratória**
- Visualizar a evolução do número de beneficiários em **gráficos de linha** para diferentes categorias (UFs, faixas etárias, tipos de contratação).
- Calcular estatísticas descritivas (média, mediana, desvios-padrão) para cada subgrupo.
- Analisar **correlações** entre variáveis (exemplo: relação entre fatores econômicos e adesão a planos de saúde).
- Realizar **decomposição de séries temporais** para observar padrões de tendência, sazonalidade e resíduos.

#### **4. Modelagem de Série Temporal**
- **Testes estatísticos** (como o teste de Dickey-Fuller aumentado - ADF) para verificar estacionariedade dos dados antes da modelagem.
- Seleção de modelos apropriados para previsão:
  - **Modelos estatísticos**: ARIMA, SARIMA, Holt-Winters, Prophet.
  - **Modelos de aprendizado de máquina**: XGBoost, LSTMs, dependendo da complexidade dos dados.
- Ajuste de hiperparâmetros e comparação do desempenho dos modelos usando métricas como **MAPE, RMSE e MAE**.

#### **5. Validação e Interpretação**
- Divisão dos dados em **conjuntos de treino e teste** (exemplo: 80% para treino, 20% para teste, com horizonte de previsão de X meses).
- Aplicação de **Validação Cruzada Temporal (Time Series Cross Validation)** para evitar overfitting.
- Interpretação dos resultados e identificação de tendências, como crescimento acelerado em determinados estados ou faixas etárias.

#### **6. Geração de Relatórios e Insights**
- Construção de **dashboards interativos** utilizando ferramentas como **Plotly Dash, Streamlit ou Power BI** para visualização dos dados.
- Geração de gráficos e tabelas para sintetizar os achados e facilitar a comunicação.
- Apresentação de recomendações baseadas nas análises para operadoras de planos de saúde, gestores públicos e órgãos reguladores.
- Exploração de possíveis impactos de políticas públicas na adesão aos planos de saúde.
